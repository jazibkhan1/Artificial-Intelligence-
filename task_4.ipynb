```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
#loading the iris dataset 
iris = load_iris()

#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.8, random_state=56)
```

```python
# Train a decision tree classifier using entropy
tree_entropy = DecisionTreeClassifier(criterion='entropy')
tree_entropy.fit(X_train, y_train)
```

```python
# Evaluate the entropy tree on the testing set
y_pred_entropy = tree_entropy.predict(X_test)
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print("Accuracy decision tree with entropy criterion:", accuracy_entropy)
```

```python
# Train a decision tree classifier using gini
tree_gini = DecisionTreeClassifier(criterion='gini')
tree_gini.fit(X_train, y_train)
```

```python
# Evaluate the gini tree on the testing set
y_pred_gini = tree_gini.predict(X_test)
accuracy_gini = accuracy_score(y_test, y_pred_gini)
print("Accuracy of decision tree with gini criterion:", accuracy_gini)
```

```python
# Visualize the decision tree for entropy
plt.figure(figsize=(16,10))
plot_tree(tree_entropy, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
# Visualize the decision tree for gini
plt.figure(figsize=(16,10))
plot_tree(tree_gini, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
